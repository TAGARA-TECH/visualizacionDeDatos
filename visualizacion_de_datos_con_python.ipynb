# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

## **Clase 1**

```python
import pandas as pd
```

```python
import pandas as pd

# Fix the URL by removing the extra space and "=df" at the end
df = pd.read_csv('https://raw.githubusercontent.com/TAGARA-TECH/visualizacionDeDatos/refs/heads/main/inmigrantes_canada.csv')

df
```

```python
df.shape
```

```python
df.info()
```

```python
# Verify the column names in the DataFrame
print(df.columns)

# Replace 'Pais' with the correct column name if it's different
df.set_index('Pais', inplace=True)
```

```python
df
```

```python

anos = list(map(str,range(1980,2014)))
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

```python
col_dict = colombia.to_dict()
col_dict
```

```python
col_dict = {'Año':colombia.index.tolist(),'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de colombianos a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.show()
```

### 08 Desafío: comparando tendencias migratorias

```python
df_comparacion = df.loc[['Brasil', 'Argentina'], anos]
```

```python
df_comparacion = df_comparacion.T
```

```python
df_comparacion.head()
```

```python
plt.plot(df_comparacion['Brasil'], label='Brasil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

## **Clase 02 - Creando figuras con Matplotlib**

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número de Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,3))

axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de colombianos hacia Canadá\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid(True)

# Call boxplot directly on the Axes object axs[1]
axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('boxplot de la Inmigración de colombianos hacia Canadá\n 1980 - 2013')
axs[1].set_xlabel('Argentina')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid(True)

plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)

axs[0, 0].plot(df.loc['Colombia',anos])
axs[0, 0].set_title('Colombia')

axs[0, 1].plot(df.loc['Brasil',anos])
axs[0, 1].set_title('Brasil')

axs[1, 0].plot(df.loc['Argentina',anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Perú',anos])
axs[1, 1].set_title('Perú')

ymin = 0
ymax = 7000
for ax in axs.flat:
    ax.set_ylim(ymin, ymax)
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Año')
    ax.set_ylabel('Número de Inmigrantes')
    ax.grid(True)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid(True)

plt.show()
```
