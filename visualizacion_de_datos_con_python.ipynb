```python

anos = list(map(str,range(1980,2014)))
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

```python
col_dict = colombia.to_dict()
col_dict
```

```python
col_dict = {'Año':colombia.index.tolist(),'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de colombianos a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Inmigracion de 4 paises latinoamericanos hacia Canada 1980 - 2013')

axs[0, 0].plot(df.loc['Colombia',anos])
axs[0, 0].set_title('Colombia')

axs[0, 1].plot(df.loc['Brasil',anos])
axs[0, 1].set_title('Brasil')

axs[1, 0].plot(df.loc['Argentina',anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Perú',anos])
axs[1, 1].set_title('Perú')

ymin = 0
ymax = 7000
for ax in axs.flat:
    ax.set_ylim(ymin, ymax)
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Año')
    ax.set_ylabel('Número de Inmigrantes')
    ax.grid(True)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de Inmigrantes')
  ax.grid(True)

plt.show()
```

```python
sudamerica_sort = sudamerica.sort_values(by='Total',ascending=True)
```

```python
colors = ['royalblue','orange','forestgreen','orchid','purple','brown','slateblue','gray','olive','navy','teal','tomato']
colores = []
for pais in sudamerica_sort.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

fig, ax=plt.subplots(figsize=(10,3))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
#ax.yaxis.set_visible(False)
#ax.tick_params(axis='x', length=0)
ax.tick_params(axis='y',which='both',size=0)

ax.barh(sudamerica_sort.index, sudamerica_sort['Total'],color=colores) # colores
ax.set_title('Sudamerica: Colombia es el segundo pais con mayor cantidad\n de Inmigrantes hacia Canada durante 1980 - 2013',fontsize=18, loc='left')
ax.set_ylabel('')
ax.set_xlabel('Número de Inmigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

for i,j in enumerate(sudamerica_sort['Total']):
  ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')

#ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#ax.tick_params(axis='x', rotation=90)

fig.savefig('/content/Sudamerica_colombia.png',dpi=300,transparent=False,bbox_inches='tight')

plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
# Crear la figura y los subgráficos
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar los espaciados entre los subgráficos
plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Agregar un título general a los subgráficos
fig.suptitle('Ventas en el periodo de enero a diciembre de 2022 en las tiendas A, B, C y D', fontsize=20)

# Lista de colores
colores = ['darkviolet', 'green', 'darkblue', 'coral']

# Convertir el diccionario de ventas a un DataFrame
df_ventas = pd.DataFrame(ventas_2022)

# Bucle para trazar y personalizar los subgráficos
# Iterar a través de los ejes planos y las columnas del DataFrame de ventas
for i, ax in enumerate(axs.flat):
    # Plotting the sales data for each store (column) against the months (index)
    ax.plot(df_ventas.index, df_ventas.iloc[:, i], color=colores[i], lw=3)
    ax.set_title(f'Ventas en la tienda {df_ventas.columns[i]}', loc='left', fontsize=16)
    ax.set_xlabel('Mes', fontsize=14)
    ax.set_ylabel('Número de ventas', fontsize=14)
    ax.tick_params(labelsize=12)
    ax.grid(color='lightgrey')

plt.show()
```
