```python
def generar_grafico(pallete):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y',which='both',size=0)
  ax = sns.barplot(data=top_10,x='Total',y=top_10.index, hue=top_10.index,orient='h',palette=pallete, legend=False)
  # Use set_title() for the title, as it accepts the 'loc' argument
  ax.set_title('Top 10 paises con mayor migración hacia canada\n1980 - 2013',
               loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  #ax.xaxis.set_tick_params(labelsize=12)
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')
  plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def generar_grafico2(pallete):
  # Establecer el estilo de Seaborn a 'white' para un fondo blanco sin cuadrícula por defecto
  sns.set_style("white")

  fig, ax = plt.subplots(figsize=(8,4))

  # Eliminar los xticklabels ya que no son necesarios en un gráfico de barras horizontal con valores anotados
  ax.set(xticklabels=[])

  ax = sns.barplot(data=top_10,x='Total',y=top_10.index, hue=top_10.index,orient='h',palette=pallete, legend=False)

  ax.set_title('Top 10 paises con mayor migración hacia canada\n1980 - 2013',
               loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')

  # Usar sns.despine() para eliminar las spines superior y derecha
  sns.despine(top=True, right=True)

  # Establecer el color de las spines (ejes) izquierda y inferior a negro
  ax.spines['left'].set_color('black')
  ax.spines['bottom'].set_color('black')

  # Eliminar las líneas de la cuadrícula vertical
  ax.xaxis.grid(False)

  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')

  plt.show()
```
